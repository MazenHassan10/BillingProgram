# Open Cases

```python
# make a data frame from  Transactions - Open Case AR Report.csv
OpenCases = pd.read_csv('../DataBase/Transactions - Open Case AR Report.csv')
OpenCases.info()
```

```python
# make date of service, date billed and Pmt Date and Discount Date columns datetime type and if empty put NaT
date_columns = ['Date of Service', 'Date Billed', 'Discount Date']
for col in date_columns:
    OpenCases[col] = pd.to_datetime(OpenCases[col], errors='coerce').dt.strftime("%m-%d-%Y")
OpenCases.info()
```

```python
OpenCases.head()
```

```python
def clean_percentage(value):
    # Remove $ sign, commas, and % symbol, then convert to float
    try:
        # If it's already a number, return it
        if isinstance(value, (int, float)):
            return float(value)
        # If it's a string, clean it and convert
        return float(str(value).replace("$", "").replace(",", "").replace("%", "").strip())
    except (ValueError, AttributeError):
        # Return None for any values that can't be converted
        return None

#apply function to Collection and Lien columns
OpenCases['Collection %'] = OpenCases['Collection %'].apply(clean_percentage)
OpenCases['Lien Reduction %'] = OpenCases['Lien Reduction %'].apply(clean_percentage)
OpenCases.head()
```

```python
#we need to recalculate the settlement Accepted Amount column in OpenCases data frame
OpenCases['Settlement Accepted'] = OpenCases['Item Charge'] - OpenCases['Item Charge'] * OpenCases['Lien Reduction %'] / 100
OpenCases['Discount Amount'] = OpenCases['Item Charge'] * OpenCases['Lien Reduction %'] / 100
OpenCases.head()
```

```python
OpenCases['Pmt Date'] = OpenCases['Pmt Date'].fillna(pd.NaT)
OpenCases['Discount Date'] = OpenCases['Discount Date'].fillna(pd.NaT)
OpenCases.head()
```

```python
OpenCases.info()
```

# Paid Cases

```python
# makre a data frame from  Transactions - Paid Case Report.csv
PaidCases = pd.read_csv('../DataBase/Transactions - Paid Case Report.csv')
PaidCases.info()
```

```python
def clean_percentage(value):
    # Remove $, commas, %, then convert to float
    try:
        return float(value.replace("$", "").replace(",", "").replace("%", "").strip()) * 100
    except:
        return None  # handles blanks or bad values safely

# Apply function
PaidCases['Settlement %'] = PaidCases['Settlement %'].apply(clean_percentage)

PaidCases.head()
```

```python
PaidCases['Discount Amount'] = PaidCases['Charge Amount'] - PaidCases['Charge Amount'] * PaidCases['Settlement %'] / 100
PaidCases.head()
```

```python
PaidCases.info()
```

```python
# make date of service, date billed and Pmt Date columns datetime type and if empty put NaT
date_columns = ['Date of Service', 'Date Billed', 'Pmt Date']
for col in date_columns:
    PaidCases[col] = pd.to_datetime(PaidCases[col], errors='coerce').dt.strftime("%m-%d-%Y")
PaidCases.head()
```

```python
PaidCases.info()
```

# Settled Cases

```python
# make a data frame from  Transactions - Settled Case Report.csv
SettledCases = pd.read_csv("../DataBase/Transactions - Settled Case Report.csv")
SettledCases.info()
```
